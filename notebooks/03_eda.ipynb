# Notebook 03 — EDA

This notebook produces the visualization artifacts used in the documentation and highlights the strongest relationships for downstream modeling.

```
import sys, pathlib
ROOT = pathlib.Path.cwd().resolve()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print('Added repository root to sys.path:', ROOT)
```

```
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import load_merged_data
from src.preprocessing import clean_dataframe
from src.visualization import plot_target_distribution
from src.eda import categorical_associations
from src.config import FIGURES_DIR

df = clean_dataframe(load_merged_data())
plot_target_distribution(df, 'status_group', 'target_distribution.png')
print('Saved target distribution figure to', FIGURES_DIR / 'target_distribution.png')
```

```
fig, ax = plt.subplots(figsize=(8,4))
sns.histplot(df['amount_tsh'].clip(upper=5000), bins=30, ax=ax)
ax.set_title('Histogram of amount_tsh')
fig.tight_layout()
fig.savefig(FIGURES_DIR / 'amount_tsh_hist.png', dpi=150)
plt.close(fig)
print('Saved amount_tsh_hist.png')
```

```
fig, ax = plt.subplots(figsize=(8,4))
sns.boxplot(data=df, x='status_group', y='amount_tsh', ax=ax)
ax.set_title('amount_tsh by status_group')
fig.tight_layout()
fig.savefig(FIGURES_DIR / 'amount_tsh_boxplot.png', dpi=150)
plt.close(fig)
print('Saved amount_tsh_boxplot.png')
```

```
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(data=df, x='water_quality', order=df['water_quality'].value_counts().index[:8], ax=ax)
ax.set_title('Water quality counts')
ax.tick_params(axis='x', rotation=45)
fig.tight_layout()
fig.savefig(FIGURES_DIR / 'water_quality_countplot.png', dpi=150)
plt.close(fig)
print('Saved water_quality_countplot.png')
```
