# Notebook 01 — Data Loading & Exploration

This notebook loads the raw feature and label tables, merges them on `id`, and profiles the data before any cleaning step.

```
import sys, pathlib
ROOT = pathlib.Path.cwd().resolve()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print('Added repository root to sys.path:', ROOT)
```

```
from src.data_loader import load_merged_data
import pandas as pd

df = load_merged_data()
df.head()
```

```
print('Shape:', df.shape)
print('Columns:', list(df.columns))
print('Dtypes:\n', df.dtypes.to_string())
print('Missing values:\n', df.isna().sum().sort_values(ascending=False).head(20).to_string())
print('Duplicates:', df.duplicated().sum())
print('Target distribution:\n', df['status_group'].value_counts(dropna=False).to_string())
```

```
numeric = df.select_dtypes(include='number')
cat = df.select_dtypes(include='object')
print('Numeric describe:\n', numeric.describe().T.to_string())
print('Categorical unique counts:\n', cat.nunique().sort_values(ascending=False).to_string())
```
