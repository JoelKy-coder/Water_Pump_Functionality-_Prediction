# Notebook 05 — Evaluation

This notebook evaluates the selected model on the held-out split, saves the final artifacts, and records the benchmark metrics for reporting.

```
import sys, pathlib
ROOT = pathlib.Path.cwd().resolve()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print('Added repository root to sys.path:', ROOT)
```

```
import json, time
from src.data_loader import load_merged_data
from src.preprocessing import clean_dataframe, encode_target
from src.model_training import train_and_select_model
from src.evaluation import compute_metrics, save_artifacts

raw_df = load_merged_data()
cleaned_df = clean_dataframe(raw_df)
encoded_df, encoder = encode_target(cleaned_df)
_, metadata, model = train_and_select_model(encoded_df)
X = encoded_df.drop(columns=['status_group'])
y = encoded_df['status_group']
pred = model.predict(X)
metrics = compute_metrics(y, pred)
print(json.dumps(metrics, indent=2))
metadata.update({'timestamp': time.strftime('%Y-%m-%d %H:%M:%S'), 'feature_list': X.columns.tolist(), 'metrics': metrics})
save_artifacts(model, encoder, metadata)
print('Artifacts saved successfully')
```
